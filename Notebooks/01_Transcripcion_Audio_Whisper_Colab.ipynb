## 3. Transcribir el audio

```python
import whisper
from pydub import AudioSegment

# Cargar el modelo base
# model = whisper.load_model("base")  # puede cambiarse por "small", "medium", "large"
model = whisper.load_model("small")  # puede cambiarse por "small", "medium", "large"
# model = whisper.load_model("medium")  # puede cambiarse por "small", "medium", "large"
# model = whisper.load_model("large")  # puede cambiarse por "small", "medium", "large"

# Reemplazar con el nombre del archivo cargado
file_name = list(uploaded.keys())[0]

# Extract audio using pydub
audio = AudioSegment.from_file(file_name)
audio_file = "audio.wav"
audio.export(audio_file, format="wav")

# Transcripción
result = model.transcribe(audio_file, language="es")

# Mostrar texto transcrito
print(result['text'])
```

## 4. Guardar la transcripción en un archivo de texto

```python
with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(result['text'])

# Descargar el archivo
from google.colab import files
files.download("transcripcion.txt")
```
