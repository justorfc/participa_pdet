```python
# ğŸ“Œ Paso 3: Extraer el audio
audio_path = "audio.wav"
!ffmpeg -i "{video_path}" -ac 1 -ar 16000 -f wav "{audio_path}" -y
```

```python
# ğŸ“Œ Paso 4: Transcribir con Whisper
import whisper
model = whisper.load_model("base")  # Puedes cambiar a "small", "medium", etc.
result = model.transcribe(audio_path)
```

```python
# ğŸ“Œ Paso 5: Guardar la transcripciÃ³n en un archivo .txt
with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("âœ… TranscripciÃ³n guardada en 'transcripcion.txt'")
```
