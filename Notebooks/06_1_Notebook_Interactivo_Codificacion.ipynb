```python
# 2. Carga de archivos `.txt`
ruta = '/content/drive/MyDrive/_Luz Stella Doctorado/Mario Chamie Mazzilli/1_Audios_Archivos .m4a/Prueba_Piloto_2'
archivos = glob.glob(os.path.join(ruta, '*.txt'))
data = []
for archivo in archivos:
    with open(archivo, 'r', encoding='utf-8') as f:
        texto = f.read().split('\n')
        for linea in texto:
            if linea.strip():
                data.append({'archivo': os.path.basename(archivo), 'fragmento': linea.strip()})
df = pd.DataFrame(data)
df['codigo_abierto'] = ''
df['categoria_axial'] = ''
df.head()
```

```python
# 3. Interfaz interactiva
index = 0
def mostrar_interfaz(i):
    clear_output(wait=True)
    if i >= len(df):
        print('✔️ Codificación finalizada.')
        return
    print(f"Fragmento {i+1}/{len(df)} del archivo: {df.loc[i, 'archivo']}")
    print("--------------------------------------------------------")
    print(df.loc[i, 'fragmento'])
    print("--------------------------------------------------------")
    codigo = widgets.Text(value='', placeholder='Ingrese código emergente', description='Código:', layout=widgets.Layout(width='100%'))
    categoria = widgets.Dropdown(options=['Participación', 'Capacidades', 'Satisfacción', 'Otro'], description='Categoría:')
    btn = widgets.Button(description='Guardar y continuar')
    def guardar(b):
        df.at[i, 'codigo_abierto'] = codigo.value
        df.at[i, 'categoria_axial'] = categoria.value
        mostrar_interfaz(i + 1)
    btn.on_click(guardar)
    display(codigo, categoria, btn)

mostrar_interfaz(index)
```

```python
# 4. Exportación de resultados
df.to_csv("codificacion_abierta_y_axial.csv", index=False)
```
